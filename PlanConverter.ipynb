```python
#region 열행 엑셀스타일
import pandas as pd

# 엑셀 파일 읽기
df = pd.read_excel('org.xlsx', sheet_name='Sheet1')

# 데이터가 있는 행만 선택 (모든 열 기준)
data_only = df.dropna(how='all')  # 모든 열이 NaN인 행 제거

# 열행 엑셀 스타일 방식으로 동적 변수 생성
def assign_excel_style_variables(df):
    """
    엑셀 스타일(열 이름 + 행 번호)로 동적 변수를 생성하고 값을 할당합니다.
    
    Parameters:
        df (DataFrame): 데이터를 포함하는 Pandas DataFrame
    """
    for row_idx in range(len(df)):
        for col_idx, col_name in enumerate(df.columns):
            # 열 이름(A, B, C, ...) 생성
            col_letter = chr(65 + col_idx)  # A=65 in ASCII
            # 변수명 생성 (열 이름 + 행 번호)
            var_name = f"{col_letter}{row_idx + 2}"  # 엑셀 행 번호 (1부터 시작)
            
            # 변수에 값 할당
            globals()[var_name] = df.iloc[row_idx, col_idx]

# 함수 실행
assign_excel_style_variables(data_only)

# 생성된 변수 출력
print("B2:", B2)  # '우리집'
print("C2:", C2)  # '년/월→'
print("D2:", D2)  # '2025'
print("E2:", E2)  # '01'

print("B3:", B3)  # '서울특별시 용산구 한강대로23길55, 9층(한강로3가)'
print("E3:", E3)  # '02-2008-9114'

print("B7:", B7)  # '에이치디씨현대산업개발(주)'
print("C7:", C7)  # '현장대리인'
print("D7:", D7)  # '강 준 형'
print("E7:", E7)  # '010-2408-2983'
#endregion
```

```python
#region 함수 모음
def 아래_표이동(n):
    """n번 아래로 이동"""
    for _ in range(n):
        hwp.TableLowerCell()

def 오른_표이동(n):
    """n번 오른쪽으로 이동"""
    for _ in range(n):
        hwp.TableRightCell()

def 셀글자전체선택():
    hwp.Cancel()
    hwp.MoveSelLineEnd()

def 첫셀로이동():
    hwp.TableColBegin()
    hwp.TableColPageUp()

def 내용삽입_표(s : str): 
    """행_열 을 문자열로 입력"""
    if s in globals():  # 동적으로 변수 값 참조
        value = globals()[s]  # 변수명으로 변수의 값 가져오기
        셀글자전체선택()
        
        # 값이 NaN인 경우 공란 처리
        if pd.isna(value):  # NaN 체크
            value = " " # 공란으로 대체
            hwp.insert_text(value)
            hwp.TableCellBlock()
            hwp.Run("TableDeleteCell")
            return
        hwp.insert_text(value)
        hwp.TableCellBlock()
    else:
        # 변수가 존재하지 않을 경우 공란 처리
        value = " "  # 또는 "값 없음" 등 원하는 기본값 설정
        셀글자전체선택()
        hwp.insert_text(value)
        hwp.TableCellBlock()
        hwp.Run("TableDeleteCell")
#endregion
```
